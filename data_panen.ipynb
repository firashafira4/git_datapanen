```python
data_panen  = {
    'lokasi1': {
        'nama_lokasi': 'Kebun A',
        'hasil_panen': {
            'padi': 1200,
            'jagung': 800,
            'kedelai': 500
        }
    },
    'lokasi2': {
        'nama_lokasi': 'Kebun B',
        'hasil_panen': {
            'padi': 1500,
            'jagung': 900,
            'kedelai': 450
        }
    },
    'lokasi3': {
        'nama_lokasi': 'Kebun C',
        'hasil_panen': {
            'padi': 1100,
            'jagung': 750,
            'kedelai': 600
        }
    },
    'lokasi4': {
        'nama_lokasi': 'Kebun D',
        'hasil_panen': {
            'padi': 1300,
            'jagung': 850,
            'kedelai': 550
        }
    },
    'lokasi5': {
        'nama_lokasi': 'Kebun E',
        'hasil_panen': {
            'padi': 1400,
            'jagung': 950,
            'kedelai': 480
        }
    }
}
```

```python
# Menampilkan seluruh data dari data panen 
for panen, data in data_panen.items():
    print(f"Lokasi: {data['nama_lokasi']}")
    for panen, hasil in data['hasil_panen'].items():
        print(f" {panen.capitalize()}: {hasil}")
    print()

# Menampilkan jumlah hasil panen jagung dari lokasi 2
jagung_lokasi2 = data_panen['lokasi2']['hasil_panen']['jagung']
print(f"Hasil panen jagung dari Lokasi 2 adalah: {jagung_lokasi2}\n")

# Menampilkan nama lokasi dari lokasi 3
lokasi3 = data_panen['lokasi3']['nama_lokasi']
print(f"Nama lokasi 3 adalah: {lokasi3}\n")

# Jumlah hasil panen padi dan kedelai setiap lokasi dalam variabel yang berbeda
hasil_padi = {key: value['hasil_panen']['padi'] for key, value in data_panen.items()}
hasil_kedelai = {key: value['hasil_panen']['kedelai'] for key, value in data_panen.items()}

print("Hasil Panen Padi:", hasil_padi)
print("Hasil Panen Kedelai:", hasil_kedelai)

# Percabangan 
for lokasi, data in data_panen.items():
    nama_lokasi = data['nama_lokasi']
    hasil_padi = data['hasil_panen']['padi']
    hasil_jagung = data['hasil_panen']['jagung']

    if hasil_padi > 1300 or hasil_jagung > 800:
        print(f"{nama_lokasi} memerlukan perhatian khusus.")
    else:
        print(f"{nama_lokasi} dalam kondisi baik.")
```
